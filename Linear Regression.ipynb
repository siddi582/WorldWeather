```python
from scipy.stats import linregress
import matplotlib.pyplot as plt
```

```python
#Avergae values for daylight and temp
temps = [18.2, 29.5, 35.9, 50.1, 58.6, 69.9, 82.1, 85.5, 73.5, 55.2, 40.3, 30.1]
daylight = [7.6, 9.2, 12.2, 13.4, 13.9, 14.2, 13.5, 11.8, 10.9, 9.9, 8.4, 7.8]
```

```python
plt.scatter(daylight,temps)
plt.xlabel('Hours of Daylight')
plt.ylabel('Average Monthly Temperate')
plt.show()
```

```python
#Produces an umber of calculations and results serving different purposes like:
#slope and intercept which builds our equation of the line, 
#the rvalue which describes how correlated 2 variables are, p value is used for hypothesis testing/statistics,
#stderr is used to describe the standard error of the model, the higher the stderror the less the model describes the data sufficiently
#As hours of daylight increases, the monthly temperature also increases
linregress(daylight,temps)
```

```python
#Save calculations and result by saving output as tuple and making it the same size as linregress output
(slope, intercept, rval, pval, stderr) = linregress(daylight,temps)
```

```python
#equation of our line; in order to add our linear line to our scatter plot see ahead
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(line_eq)
```

```python
#Multiply each daylight values by slope and intercept using list comphrension
regress_values = [(day * slope + intercept) for day in daylight]
```

```python
plt.scatter(daylight, temps)
plt.plot(daylight, regress_values)
plt.xlabel('Hours of Daylight')
plt.ylabel('Average Monthly Temperature')
plt.show()
```
